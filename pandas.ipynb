```python
#df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison')]
#df.loc[(df['Type 1']=='Grass') | (df['Type 2']=='Poison')]  #  | =or, &= and
df_filtered=df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison') & (df['HP']>70)] #esto se  podria guardar como un documento ya filtrado
df_filtered.reset_index() # reset el anterior indice
df_filtered.reset_index(drop=True) #quitar el verdadero indice
```
