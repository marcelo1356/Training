```python
import numpy as np
import matplotlib.pyplot as plt

alpha=2.43
zf=0.33
q=0.79
xd=0.95
xw=0.05
R=3.5

x=np.linspace(0,1,101)
yd=x #対角線
ye=(alpha*x)/((alpha-1)*x+1)#気液平衡線
xi=(zf/(1-q)-xd/(R+1))/(R/(R+1)+q/(1-q))
def yconcentrator(x):
    return R*x/(R+1)+xd/(R+1) ##濃縮部操作線
xc=np.linspace(xd,xi)
yc=yconcentrator(xc)
def yqsen(x):
    return -q*x/(1-q)+zf/(1-q) ##濃縮部操作線

xq=np.linspace(zf,xi)
yq=yqsen(xq)
m=(yqsen(xi)-xw)/(xi-xw)
def yrect(x):

    return m*x-m*xw+xw
xr=np.linspace(xw,xi)
yr=yrect(xr)
plt.plot(x,yd) #対角線
plt.plot(x,ye) #気液平衡線
plt.plot(xc,yc) #濃縮部操作線
plt.plot(xq,yq) #濃縮部操作線
plt.plot(xr,yr) 
plt.show()

```

```python
i=1
py=[0]
px=[0]
yi=yqsen(xi)
py[0]=0.95
px[0]=0.95
while px[i-1]>xi and py[i-1]>yi:    
    py.append(0)
    py[i]=py[i-1]  
    px.append(0)
    px[i]=py[i]/(alpha-(alpha-1)*py[i])
    
    px.append(0)
    px[i+1]=px[i]
    py.append(0)
    py[i+1]=yconcentrator(px[i+1])
    i=i+2
py[-1]=yrect(px[-1])
a=len(px)
k=len(px)
while px[k-1]>xw and py[k-1]>xw:    
    py.append(0)
    py[k]=py[k-1]  
    px.append(0)
    px[k]=py[k]/(alpha-(alpha-1)*py[k])
    
    px.append(0)
    px[k+1]=px[k]
    py.append(0)
    py[k+1]=yrect(px[k+1])
    k=k+2
py[-1]=px[-1]

plt.plot(x,yd) #対角線
plt.plot(x,ye) #気液平衡線
plt.plot(xc,yc) #濃縮部操作線
plt.plot(xq,yq) #濃縮部操作線
plt.plot(xr,yr) 
plt.plot(px,py)
plt.show()
print((len(px)-1)/2)
print((a-1)/2)
```
